```python
#创建一个图片
plt.plot(np.random.randn(30).cumsum(), 'bo--')
plt.show()
```

```python
#创建多个窗口图
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
ax3.plot(np.random.randn(500).cumsum(), 'b--')

plt.show()
```

```python
#创建多个窗口图并设置图间距为0
fig, axes = plt.subplots(2, 3, sharex=True, sharey=True)  #创建多窗口图
for i in range(2):
    for j in range(3):
        axes[i, j].hist(np.random.randn(500), bins=50, color='b', alpha=0.5)
        
plt.subplots_adjust(wspace=0, hspace=0) #调整各个图之间的间距

plt.show()
```

```python
#xlim--x轴绘图范围
#说明轴的自定义，创建一个随机漫步图
fig = plt.figure();
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
plt.show()
```

```python
#说明轴的自定义，创建一个随机漫步图-修改X轴的标签刻度标题字体大小旋转角度等
fig = plt.figure();
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])  #xticks修改X轴的刻度
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=10, fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
plt.show()
```

```python
#创建有三条线和图例的示例：图例设置方法-legends
fig = plt.figure();
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'b', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'g--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'r.', label='three')
ax.text(50, 1, 'hello world!', family='monospace', fontsize=10)
ax.legend(loc='best')
plt.show()
```

```python
#使用注释和箭头示例
from datetime import datetime

fig = plt.figure();
ax = fig.add_subplot(1, 1, 1)
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='b-')

crisis_data = [
 (datetime(2007, 10, 11), 'Peak of bull market'),
 (datetime(2008, 3, 12), 'Bear Stearns Fails'),
 (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
 ax.annotate(label, xy=(date, spx.asof(date) + 50),
 xytext=(date, spx.asof(date) + 200),
 arrowprops=dict(facecolor='green'),
 horizontalalignment='left', verticalalignment='top')
# Zoom in on 2007-2010
ax.set_xlim(['1/1/2006', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in 2008-2009 financial crisis')

plt.show()
```

```python
#图形示例程序 矩形 三角形 椭圆
fig = plt.figure();
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.25, 0.75), 0.6, 0.3, color='b', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)

plt.show()

#plt.savefig('figpath.svg')
#plt.savefig('figpath.png', dpi=400, bbox_inches='tight')

#from io import StringIO
#buffer = StringIO()
#plt.savefig(buffer)
#plot_data =buffer.getvalue()
#print plot_data
```
